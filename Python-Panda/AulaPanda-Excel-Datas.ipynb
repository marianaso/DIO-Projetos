# **Excel**

```python
import pandas as pd
```

```python
##Leitura de cada arquivo
df1 = pd.read_excel("/content/drive/MyDrive/datasets/Aracaju.xlsx")
df2 = pd.read_excel("/content/drive/MyDrive/datasets/Fortaleza.xlsx")
df3 = pd.read_excel("/content/drive/MyDrive/datasets/Natal.xlsx")
df4 = pd.read_excel("/content/drive/MyDrive/datasets/Recife.xlsx")
df5 = pd.read_excel("/content/drive/MyDrive/datasets/Salvador.xlsx")
```

```python
##juntar todos os arquivos
df = pd.concat([df1,df2,df3,df4,df5])
df.head()
df.sample(10) ##Pega uma amostra randomica desse banco de dados, em () o N 
df.dtypes
```

```python
df["LojaID"] = df["LojaID"].astype("object") ##Modifica o tipo da coluna LojaID, não é obrigatorio
df.isnull() ##Pesquisa se os valores são nulos
df.isnull().sum() ##Consulta quantos valores nulos há em cada coluna, valores faltantes no caso

df['Vendas'].fillna(0, inplace=True) ##Completa as linhas vazias com 0 ou o parametro passado
df.dropna(inplace=True) ##Apaga as linhas com valores nulos
df.dropna(subset=['Vendas'], inplace=True) ##Apaga as linhas com valores nulos com base na coluna Vendas
df.dropna(how="all", inplace=True) ##Remove linhas que estejam com valores faltantes em todas as colunas
```

```python
##Criando uma nova coluna - RECEITA (o quanto entrou)
df["Receita"] = df["Vendas"].mul(df["Qtde"])
df.head()
```

```python
df['Receita'].max()
```

```python
df['Receita'].min()
```

```python
df.nlargest(2, "Receita") ## Retorna as 2 maiores receitas
```

```python
df.groupby("Cidade")["Receita"].sum() ##Soma a receita por cidade
```

```python
df.sort_values("Receita", ascending=False).head(10) ##Ordena conforme o parametro passado
df.sort_values("Receita") # Printa tudo e ascendente
```

## **Datas**

```python
df["Data"] = df["Data"].astype("int64")
##Muitas vezes quando lemos uma tabela os dados de data não estão no formato data
```

```python
df['Data'].dt.year ##Filtra todos os anos das datas
df.groupby(df['Data'].dt.year)['Receita'].sum() ##Agrupa a soma das receitas por ano
```

```python
df.groupby(df['Data'].dt.month)['Receita'].sum() ##Agrupa a soma das receitas por mes
```

```python
df.groupby(df['Cidade'])['Receita'].max() ##Mostra o maximo de receita por Cidade
df.groupby(df['Cidade'])['Receita'].min() ##Mostra o minimo de receita por Cidade
df.groupby(df['Cidade'])['Receita'].sum() ##Mostra a soma de receita por Cidade
```

```python
df['AnoVenda'] = df['Data'].dt.year ##Cria uma coluna com o ano da venda
```

```python
df['MesVenda'], df['DiaVenda'] = (df['Data'].dt.month, df['Data'].dt.day) ##Cria uma coluna com o ano da venda
```

```python
##Diferença entre datas
df['DiferencaDias'] = df['Data'] - df['Data'].min() ##Não consegue subtrair duas colunas pois só há uma coluna de data
df.sample(8)
```

```python
## Coluna trimestre
df['Trimestre'] = df['Data'].dt.quarter
```

```python
## Filtras vendas
##.loc = localizaçao
vendasMarco2019 = df.loc[(df['Data'].dt.year == 2019) & (df['Data'].dt.month == 3)]
vendasMarco2019
```
