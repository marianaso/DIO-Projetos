```python
df = pd.read_excel("/content/AdventureWorks.xlsx")
```

```python
df.shape ##Numero de linhas e colunas
```

```python
df.dtypes
```

```python
df['Valor Venda'].sum() ##Receita Total
```

```python
df['Custo'] = df['Custo Unitário'] * df['Quantidade'] #Custo produto
```

```python
#Qual o custo total?
round(df['Custo'].sum(), 2)
```

```python
#Qual o lucro?
df['Lucro'] = df['Valor Venda'] - df['Custo']
```

```python
#Lucro total
round(df['Lucro'].sum(), 2)
```

```python
#Tempo de envio
df['Tempo Envio'] = df['Data Envio'] - df['Data Venda']
```

```python
##Media de tempo de envio para cada marca
df['Tempo Envio'] = df['Tempo Envio'].dt.days
```

```python
df.groupby(df['Marca'])['Tempo Envio'].mean()
```

```python
#Verificar se há valores faltantes/nulos
df.isnull().sum()
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
#Lucro por ano e marca
df.groupby([df['Data Venda'].dt.year, 'Marca'])['Lucro'].sum()
```

```python
#Resetando o index, traz em formato colunar
lucroAno = df.groupby([df['Data Venda'].dt.year, 'Marca'])['Lucro'].sum().reset_index()
lucroAno
```

```python
#Total de Produtos Vendidos
gb1 = df.groupby('Produto')['Quantidade'].sum().sort_values(ascending = False)
```

```python
## GRAFICO COM OS PRODUTOS VENDIDOS
gb1.plot.barh(title = 'Total Produtos Vendidos')
plt.xlabel('Quantidade')
plt.ylabel('Produto');
```

```python
## GRAFICO LUCRO POR ANO
df.groupby(df['Data Venda'].dt.year)['Lucro'].sum().plot.bar(title='Lucro por Ano')
plt.xlabel('Ano')
plt.ylabel('Receita');
```

```python
## Filtra para as vendas de 2009
df2009 = df[df['Data Venda'].dt.year== 2009]
```

```python
## Grafico com o lucro por mes em 2009
df2009.groupby(df2009['Data Venda'].dt.month)['Lucro'].sum().plot(title='Lucro x Mes')
plt.xlabel('Mes')
plt.ylabel('Lucro');
```

```python
# Grafico do lucro por marca em 2009
df2009.groupby('Marca')['Lucro'].sum().plot.bar(title='Lucro x Marca')
plt.xlabel('Marca')
plt.ylabel('Lucro')
plt.xticks(rotation = 'horizontal');
```

```python
#Lucro por Classe
df2009.groupby('Classe')['Lucro'].sum().plot.bar(title='Lucro x Marca')
plt.xlabel('Classe')
plt.ylabel('Lucro')
plt.xticks(rotation = 'horizontal');
```
